```python
import csv
import re
import string
import json
import numpy as np
from ast import literal_eval

import openpyxl

file_needs = 'needs.xlsx'

output_file_pair_need = 'need_pairs.csv'
out_file_pair_words = 'pair_words.csv'


pair_need_columnas = [
    'need_id',
    'pair_id',
]

pair_words_columnas = [
    'pair_id',
    'palabra_inicio',
    'palabra_fin',
]
```

```python
book = openpyxl.load_workbook(file_needs)

sheet = book.active

pair_needs = list()
pair_words = list()

pair_ids = 0

for row in sheet.iter_rows(min_row=2):
    need_id = int(row[0].value)
    
    role_token = row[4].value
    if role_token == None or role_token == 'NR':
        role_token = '[]'
        
    array_token = np.array(literal_eval(role_token))    
    
    pairs = []
    
    for i in range(0,len(array_token) - 1):
        pairs.append((array_token[i], array_token[i+1]))
   
    for pair in pairs:
        pair_id = pair_ids
        pair_words.append({
            'pair_id': pair_id,
            'palabra_inicio': pair[0],
            'palabra_fin': pair[1],
        })
        
        pair_needs.append({
            'pair_id': pair_id,
            'need_id': need_id,
        })
        
        pair_ids += 1
        
with open(output_file_pair_need, 'w') as new_file:
    writer = csv.DictWriter(new_file, fieldnames = pair_need_columnas)
    writer.writeheader()
    for data in pair_needs:
        writer.writerow(data)
        
with open(out_file_pair_words, 'w') as new_file:
    writer = csv.DictWriter(new_file, fieldnames = pair_words_columnas)
    writer.writeheader()
    for data in pair_words:
        writer.writerow(data)        
```
